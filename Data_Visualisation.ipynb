```python
#Histogram

import matplotlib.pyplot as plt
import numpy as np

data = np.array([12, 22, 13, 26, 32, 26])

plt.hist(data, color = 'lightgreen', edgecolor='black')

plt.title('Histogram Graph')
plt.xlabel('Value Range')
plt.ylabel('Frequency')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Creating a pandas DataFrame

data = {
    'Category' : ['A', 'B', 'C', 'D', 'E'],
    'Value' : [23, 45, 56, 78, 33],
    'Score' : [89, 92, 85, 80, 88]
}

df = pd.DataFrame(data)

#Set Seaborn style
sns.set_style('darkgrid') #Option: 'white', 'dark', 'ticks', 'darkgrid', 'whitegrid'

#Set Seaborn color palette
sns.set_palette('muted') #Options: 'deep', 'muted', 'pastel', 'dark', 'colorblind'

#Set Seaborn plot
sns.barplot(x = 'Category', y = 'Value', data = df)

#Adding labels
plt.xlabel('Category', fontsize=12, fontweight='bold')
plt.ylabel('Value', fontsize = 12, fontweight = 'bold')
plt.title('Bar plor using pandas DataFrame', fontsize= 14, fontweight = 'bold')

plt.show()
```

```python
#Data Visulization using Bokeh
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool

output_notebook()

p = figure(title= "Simple Bokeh Plot", x_axis_label = 'X', y_axis_label = 'Y')

p.line([1,2,3,4,5], [2,4,6,8,10], line_width = 2, color='blue')

p.scatter([1,2,3,4,5], [2,4,6,8,10], size = 10, color = 'red')

#Add hover tool
hover = HoverTool()
hover.tooltips = [('X', '$x'), ('Y', '$y')]
p.add_tools(hover)

show(p)
```
